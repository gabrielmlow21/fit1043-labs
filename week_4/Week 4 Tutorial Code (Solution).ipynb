### Visualizing the Data

```python
df['SBP'].head()
```

```python
# Practise 3
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(df['HR'],df['SBP'])
plt.xlabel('Heart Rate (HR)')
plt.ylabel('Systolic Blood Pressure (SBP)')
plt.title("HR vs SBP");
```

```python
# Practise 3
%matplotlib inline
import matplotlib.pyplot as plt
plt.scatter(df['HR'],df['SBP'],c=df['DBP'],cmap='hot')
plt.xlabel('Heart Rate (HR)')
plt.ylabel('Systolic Blood Pressure (SBP)')
plt.title("HR vs SBP (hot)")
plt.colorbar()
```
